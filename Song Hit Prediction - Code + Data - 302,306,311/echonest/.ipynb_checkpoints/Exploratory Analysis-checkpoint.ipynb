```python
import csv
import pandas as pd
import numpy as np
import math
from sklearn import cluster

```

```python
dfs = []
for y in xrange(1970,1990):
    read_file_name = 'param/' + str(y) + '_param.csv'
    df = pd.read_csv(read_file_name)
    dfs = dfs + [df]
    
data_param = pd.concat(dfs)

data_param = data_param.dropna()

data_param = data_param.set_index([range(1,len(data_param)+1)])




       
    
```

```python

data = data_param[['acousticness', 'danceability', 'duration', 'energy', 'key', 'liveness', 'loudness', 'mode', 'speechiness', 'tempo', 'time_signature', 'valence']]
print data
```

```python
k = 7
kmeans = cluster.KMeans(n_clusters=k)
cluster = kmeans.fit_predict(data)
```

```python
data_param['cluster'] = cluster
```
