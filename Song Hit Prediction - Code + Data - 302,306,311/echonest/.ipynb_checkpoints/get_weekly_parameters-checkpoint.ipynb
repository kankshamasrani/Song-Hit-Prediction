```python
start_date = datetime.date(1961,12,16)
end_date = datetime.date(1961,12,30)
week = datetime.timedelta(weeks=1)

while start_date != end_date:
        
    write_file_name = 'weekly/param/' + start_date.isoformat() + '_param.csv'

    fp = open(write_file_name,'w')
    f = csv.writer(fp, delimiter=',')

    headings = ['rank', 'song_name', 'artist', 'latitude', 'longitude', 'acousticness', 'danceability', 'duration', 'energy', 'key', 'liveness', 'loudness', 'mode', 'speechiness', 'tempo', 'time_signature', 'valence']
    f.writerow(headings)
    
    read_file_name = 'weekly/' + start_date.isoformat() + '.csv'
    
    rank = 0
    
    with open(read_file_name,'rb') as csvfile:        
        
        song_reader = csv.reader(csvfile, delimiter = ',')
        
        prev_file_name = 'weekly/param/' + (start_date-week).isoformat() + '_param.csv'
        prev_dataframe = pd.read_csv(prev_file_name)
        
        print read_file_name
        
        song_data = pd.DataFrame(columns = prev_dataframe.columns)
        
        for row in song_reader:
            if not rank:
                rank = rank + 1
                continue
                
            try:                   
                row[1],row[2] = row[2],row[1]
                
                print row[1],row[2]
                
                if not prev_dataframe[(prev_dataframe['song_name']==row[2]) & (prev_dataframe['artist']==row[1])].empty:
                    song_df = prev_dataframe[(prev_dataframe['song_name']==row[2]) & (prev_dataframe['artist']==row[1])]
                    song_data.append(song_df)
                    print 'found prev'
                    
                else:                    
                    print row[1],row[2], 'search echonest'
                    rkp_results = song.search(artist=row[1], title=row[2])
                    lat = rkp_results[0].artist_location['latitude']
                    lng = rkp_results[0].artist_location['longitude']
                    #print lat,lng
                    song_series = pd.Series([rank, row[2], row[1], lat, lng, rkp_results[0].audio_summary['acousticness'], rkp_results[0].audio_summary['danceability'], rkp_results[0].audio_summary['duration'], rkp_results[0].audio_summary['energy'], rkp_results[0].audio_summary['key'], rkp_results[0].audio_summary['liveness'], rkp_results[0].audio_summary['loudness'], rkp_results[0].audio_summary['mode'], rkp_results[0].audio_summary['speechiness'], rkp_results[0].audio_summary['tempo'], rkp_results[0].audio_summary['time_signature'], rkp_results[0].audio_summary['valence']])
                    f.writerow() 
#                     rank = rank + 1

            except:
                f.writerow([rank, row[2], row[1]])
                rank = rank + 1
                print 'error'
                continue           
        
                
    start_date = start_date + week
    
    curr_date = start_date
```
